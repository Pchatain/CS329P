#Function

```python
def initialize_layers(input_dim, max_depth):
  Layers = dict()
  Layers['W'] = torch.randn((input_dim, ((2**max_depth)-1)), dtype=torch.float64, device=device, requires_grad=True)
  Layers['b'] = torch.randn((((2**max_depth)-1), ), dtype=torch.float64, device=device, requires_grad=True)
  Layers['p'] = torch.randn((2**max_depth, ), dtype=torch.float64, device=device, requires_grad=True)
  m = dict()
  v = dict()
  g = dict()
  mh = dict()
  vh = dict()
  for namee in Layers.keys():
    m[namee] = torch.zeros_like(Layers[namee])
    v[namee] = torch.zeros_like(Layers[namee])
    g[namee] = torch.zeros_like(Layers[namee])
    mh[namee] = torch.zeros_like(Layers[namee])
    vh[namee] = torch.zeros_like(Layers[namee])

  return Layers, m, v, g, mh, vh
```

```python
def forward(input, Layers, max_depth=None, requires_grad=True):
  if requires_grad:
    temp = torch.sigmoid(input@Layers['W']+Layers['b'])
  else:
    temp = torch.sigmoid(input@Layers['W'].data+Layers['b'].data)
  dist = None
  if max_depth == None:
    max_depth = int(np.round(np.log(Layers['W'].shape[1]+1)/np.log(2)))
  for i in range(max_depth):
    if dist == None:
      dist = torch.cat((temp[:,2**i - 1: 2**(i+1) - 1].unsqueeze(-1), (1-temp[:,2**i - 1: 2**(i+1) - 1].unsqueeze(-1))), dim=2).reshape(input.shape[0],-1).unsqueeze(-1).repeat(1,1,2**(max_depth-1-i)).reshape(input.shape[0],-1)
    else:
      dist *= torch.cat((temp[:,2**i - 1: 2**(i+1) - 1].unsqueeze(-1), (1-temp[:,2**i - 1: 2**(i+1) - 1].unsqueeze(-1))), dim=2).reshape(input.shape[0],-1).unsqueeze(-1).repeat(1,1,2**(max_depth-1-i)).reshape(input.shape[0],-1)
  if requires_grad:
    score = dist@Layers['p']
  else:
    score = dist@Layers['p'].data
  prob = torch.sigmoid(score)
  return prob
```

```python
#Note that the output is torch
def find_loss(prob, output):
  loss = - (torch.log(prob)@output + torch.log(1-prob)@(1-output)) / output.shape[0]
  return loss

#Note that the output is torch
def find_accuracy(prob, output):
  acc = torch.mean(((2*(prob>=0.5)-1)*(2*(output>=0.5)-1)+1)/2)
  return acc
  
def find_auc (prob, output):
  roc_auc_score((output>=0.5),prob)
  return auc
```

```python
def print_progress(Layers, t_train_x, t_train_y, t_valid_x, t_valid_y, max_depth=None, epoch_index=None, verbose=True):
  #Eval Train
  prob = forward(t_train_x, Layers, max_depth=max_depth, requires_grad=False)
  train_loss = float(find_loss(prob, t_train_y))
  train_acc = float(find_accuracy(prob, t_train_y))
  if verbose:
    if epoch_index != None:
      print('\nEpoch ' + str(epoch_index) + ' Summary:')
    print("Train:")
    print('Loss: ' + str(train_loss))
    print('Accuracy: ' + str(int(10000*train_acc)/100) + '%')

  #Eval Valid
  prob = forward(t_valid_x, Layers, max_depth=max_depth, requires_grad=False)
  val_loss = float(find_loss(prob, t_valid_y))
  val_acc = float(find_accuracy(prob, t_valid_y))
  if verbose:
    print("Valid:")
    print('Loss: ' + str(val_loss))
    print('Accuracy: ' + str(int(10000*val_acc)/100) + '%')

  return train_loss, train_acc, val_loss, val_acc
```

```python
def soft_decision_tree_train(max_depth, t_train_x, t_train_y, t_valid_x, t_valid_y, mask=None, pretrained_Layers=None, step_size=1e-3, reg_l=1e-3, minibatch_size=128, num_epoch=20, num_step_per_epoch=320, beta1=0.9, beta2=0.999, epsilon=1e-8, verbose=False, graphing=False):
  #Initialization phrase
  t = 0
  
  Layers, m, v, g, mh, vh = initialize_layers(t_train_x.shape[1], max_depth)

  if pretrained_Layers != None:
    Layers = pretrained_Layers

  #Record
  loss_array = np.zeros(0)
  acc_array = np.zeros(0)
  val_loss_array = np.zeros(0)
  val_acc_array = np.zeros(0)

  #Place Mask if applicable
  if mask != None:
    Layers['W'].data *= mask

  if verbose:
    print("\nStart training\n")

    print('# of Epochs: ' + str(num_epoch))
    print('# of Steps per Epoch: ' + str(num_step_per_epoch))
    print('Minibatch Size: ' + str(minibatch_size))

  for epoch_index in range(num_epoch):
    #Print progress
    train_loss, train_acc, val_loss, val_acc = print_progress(Layers, t_train_x, t_train_y, t_valid_x, t_valid_y, max_depth=max_depth, epoch_index=epoch_index, verbose=verbose)
    loss_array= np.append(loss_array, train_loss)
    acc_array = np.append(acc_array, train_acc)
    val_loss_array= np.append(val_loss_array, val_loss)
    val_acc_array = np.append(val_acc_array, val_acc)
    for step_index in range(num_step_per_epoch):
      #Minibatch
      index = torch.randperm(t_train_x.shape[0])[:minibatch_size]

      #Forward
      prob = forward(t_train_x[index], Layers, max_depth=max_depth, requires_grad=True)
      loss = find_loss(prob, t_train_y[index])

      #Reg
      reg = reg_l * torch.mean(torch.abs(Layers['W']))
      reg_w = 1e-1
      reg_atten = 1e-2
      for i in range(max_depth):
        reg_w *= reg_atten
        vector_size = np.int(np.power(2, i))
        num_pairs = np.int(np.power(2, (max_depth-i-1)))
        for ii in range(num_pairs):
          reg += reg_w * torch.linalg.norm(Layers['p'][2*ii*vector_size:2*ii*vector_size+vector_size] - Layers['p'][2*ii*vector_size+vector_size:2*ii*vector_size+2*vector_size])

      #Backward
      loss.backward()
      reg.backward()
      
      #Update
      t += 1
      for namee in Layers.keys():
        g[namee] = Layers[namee].grad
        if (namee == 'W')*(mask!=None):
          g[namee] = g[namee]*mask
        m[namee] = beta1 * m[namee] + (1 - beta1) * g[namee]
        v[namee] = beta2 * v[namee] + (1 - beta2) * torch.square(g[namee])
        mh[namee] = m[namee] / (1 - np.power(beta1, t)) + (1 - beta1) * g[namee] / (1 - np.power(beta1, t))
        vh[namee] = v[namee] / (1 - np.power(beta2, t))
        Layers[namee].data -= step_size * mh[namee] / (torch.sqrt(vh[namee]) + epsilon)
        Layers[namee].grad.data.zero_()
  train_loss, train_acc, val_loss, val_acc = print_progress(Layers, t_train_x, t_train_y, t_valid_x, t_valid_y, max_depth=max_depth, epoch_index=epoch_index, verbose=verbose)
  loss_array= np.append(loss_array, train_loss)
  acc_array = np.append(acc_array, train_acc)
  val_loss_array= np.append(val_loss_array, val_loss)
  val_acc_array = np.append(val_acc_array, val_acc)

  if verbose:
    print("\nFinish training\n")

  if graphing:
    plt.imshow(np.abs(Layers["W"].cpu().detach().numpy()), cmap='Greys')
    plt.title('Absolute Value Intensity of All Nodes')
    plt.colorbar()
    plt.show()

    plt.plot(100*(acc_array), label='Train')
    plt.plot(100*(val_acc_array), label='Validation')
    plt.title('Accuracy')
    plt.ylabel('Accuracy Percentage')
    plt.xlabel('Epochs')
    plt.legend()
    plt.show()

    plt.plot(loss_array, label='Train')
    plt.plot(val_loss_array, label='Validation')
    plt.title('Cross Entropy Loss')
    plt.ylabel('Loss')
    plt.xlabel('Epochs')
    plt.legend()
    plt.show()
  
  return Layers
```

```python
def create_mask(W):
  mask = torch.zeros_like(W, device=device, requires_grad=False)
  mask[torch.argmax(torch.abs(W), dim=0),torch.arange(W.shape[1])] = 1
  return mask
```

```python
def find_roc_auc_after_training(t_valid_x, t_valid_y, Layers, threholsd_part=1e-2):
  prob = forward(t_valid_x, Layers, max_depth=None, requires_grad=False).cpu().detach().numpy()
  output = t_valid_y.cpu().detach().numpy()
  return find_auc(prob, output, threholsd_part=1e-2)
```

```python
def predict(prob, threshold=0.5):
  return (prob>=threshold)

def find_TPR_FPR(prob, output, threholsd_part=1e-2):
  TRPs = []
  FPRs = []

  for i in range(1, int(1 / threholsd_part)):
    threshold = threholsd_part*i
    temp = predict(prob, threshold=threshold) + 2*predict(output, threshold=0.5)
    FP = np.sum(temp==1)
    TP = np.sum(temp==3)
    FN = np.sum(temp==2)
    TN = np.sum(temp==0)
    TPR = TP / (TP+FN)
    FPR = FP / (FP+TN)
    TRPs.append(TPR)
    FPRs.append(FPR)

  FPRs = np.array(FPRs)
  TRPs = np.array(TRPs)

  return TRPs, FPRs

def find_auc(prob, output, epsilon=1e-5, plott=True, threholsd_part=1e-2):
    #threholsd_part = 0.1
    u = 1
    l = 0
    auc = 0
    while u-l >= epsilon:
      threholsd_part *= 1e-1
      
      TRPs, FPRs = find_TPR_FPR(prob, output, threholsd_part=threholsd_part)

      difference = np.append(1, FPRs) - np.append(FPRs, 0)
      u = difference[:-1]@TRPs
      l = difference[1:]@TRPs
                                                  
      auc = (difference[:-1]@TRPs + difference[1:]@TRPs)/2
    if plott:
      plt.fill_between(FPRs, TRPs, alpha=0.5, color='yellow')
      plt.plot([0,1],[0,1], '--', color='green')
      plt.xlabel("FPR")
      plt.ylabel("TPR")
      plt.title("AUC-ROC Curve")

      plt.show()

    return auc
```

#Train SSDT Function

```python
def ssdt_train(max_depth, t_train_x, t_train_y, t_valid_x, t_valid_y, step_size=1e-3, reg_l_1=1e-2, reg_l_2=5e-4, minibatch_size=128, num_epoch=20, num_step_per_epoch=500, verbose=False, graphing=False):
  Layers = soft_decision_tree_train(max_depth, t_train_x, t_train_y, t_valid_x, t_valid_y, step_size=step_size, reg_l=reg_l_1, minibatch_size=minibatch_size, num_epoch=num_epoch, num_step_per_epoch=num_step_per_epoch, verbose=verbose, graphing=graphing)
  mask = create_mask(Layers['W'].data)
  Layers['W'].data *= mask
  Layers = soft_decision_tree_train(max_depth, t_train_x, t_train_y, t_valid_x, t_valid_y, mask=mask, pretrained_Layers=Layers, step_size=step_size, reg_l=reg_l_2, minibatch_size=minibatch_size, num_epoch=num_epoch, num_step_per_epoch=num_step_per_epoch, verbose=verbose, graphing=graphing)
  return Layers
```

```python
temp = torch.sigmoid(t_train_x@Layers['W'].data + Layers['b'].data)
dist = None
for i in range(max_depth):
  if dist == None:
    dist = torch.cat((temp[:,2**i - 1: 2**(i+1) - 1].unsqueeze(-1), (1-temp[:,2**i - 1: 2**(i+1) - 1].unsqueeze(-1))), dim=2).reshape(t_train_x.shape[0],-1).unsqueeze(-1)\
    .repeat(1,1,2**(max_depth-1-i)).reshape(t_train_x.shape[0],-1)
  else:
    dist *= torch.cat((temp[:,2**i - 1: 2**(i+1) - 1].unsqueeze(-1), (1-temp[:,2**i - 1: 2**(i+1) - 1].unsqueeze(-1))), dim=2).reshape(t_train_x.shape[0],-1).unsqueeze(-1).\
    repeat(1,1,2**(max_depth-1-i)).reshape(t_train_x.shape[0],-1)

Big_prob_array = torch.zeros((dist.shape[0], (dist.shape[1]-1)), device=device, dtype=torch.float64)
for i in range(max_depth):
  Big_prob_array[:,2**(i)-1:2**(i+1)-1] = torch.sum(dist.reshape(2**i * t_train_x.shape[0],2**(max_depth-i)), dim=1).unsqueeze(dim=-1).reshape(2**i,t_train_x.shape[0]).T
```
